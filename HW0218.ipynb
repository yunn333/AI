ÊàëÊåë‰∫ÜÂπæÂÄãÊØîËºÉÁâπÂà•ÁöÑÂáΩÊï∏ÂúñÂΩ¢ <br>
ËÄå‰∏î‰∏ªÈ°åÈÉΩËàáÂ§ßËá™ÁÑ∂ÊúâÈóúÔºÅ

- Ëù∏ÁâõÊõ≤Á∑ö üêå ‚Üí Êï∏Â≠∏ËàáËá™ÁÑ∂ÁöÑÁµêÂêà

ÂÖàÂæûËä±ÊúµÊõ≤Á∑öÈñãÂßãÂÅöËµ∑ <br>
ÂÖ∂ÂÖ¨ÂºèÁÇ∫ $r = a sin(k\theta)$ Êàñ $r = a cos(k\theta)$ <br>
Áï∂ $k$ ÁÇ∫Â•áÊï∏ÊôÇÊúÉÊúâ $k$ ÁâáËä±Áì£ <br>
Áï∂ $k$ ÁÇ∫ÂÅ∂Êï∏ÊôÇÂâáÊúÉÊúâ $2k$ ÁâáËä±Áì£

```python
# ÂèÉÊï∏Ë®≠ÂÆö
theta = np.linspace(0, 2 * np.pi, 1000)
k = 10  # Ëä±Áì£Êï∏ÁöÑÊéßÂà∂ÂèÉÊï∏
r = np.cos(k * theta)

# ËΩâÊèõÁÇ∫ x, y
x = r * np.cos(theta)
y = r * np.sin(theta)

# Áπ™Ë£ΩËä±ÊúµÊõ≤Á∑ö
plt.figure(figsize=(6,6))
plt.plot(x, y, color='pink', linewidth=2)

plt.title("Flower Curve", fontsize=14)
plt.xticks([])
plt.yticks([])
plt.grid(False)

plt.show()
```

ÂÜç‰æÜÊòØËù¥Ëù∂Êõ≤Á∑ö <br>
ÂÖ∂ÂÖ¨ÂºèÂ¶Ç‰∏ã <br><br>
$ x = sin(t)(e^{cos(t)} - 2cos(4t) - sin^5(\frac{t}{12})) \\
y = cos(t)(e^{cos(t)} - 2cos(4t) - sin^5(\frac{t}{12})) $ <br><br>
ÂÖ∂‰∏≠ $t$ ‰ªãÊñº $‚àí12\pi$ Âà∞ $12\pi$ ‰πãÈñì

```python
# Ë®≠ÂÆöÂèÉÊï∏ÁØÑÂúç
t = np.linspace(-12*np.pi, 12*np.pi, 10000)

# Ë®àÁÆóËù¥Ëù∂Êõ≤Á∑öÁöÑ x, y
x = np.sin(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12)**5)
y = np.cos(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12)**5)

# Áπ™Ë£ΩËù¥Ëù∂Êõ≤Á∑ö
plt.figure(figsize=(8,6))
plt.plot(x, y, color='purple', linewidth=2)

# Ë®≠ÂÆöÊ®ôÈ°å
plt.title("Butterfly Curve", fontsize=14)
plt.xticks([])
plt.yticks([])
plt.grid(False)

# È°ØÁ§∫ÂúñÂΩ¢
plt.show()
```

ÊúÄÂæåÊòØËù∏ÁâõÊõ≤Á∑ö <br>

$r = e^{0.2t} \\

‰∏çÈÅé‰∏ªË¶ÅÂè™ÊúÉÁî¢ÁîüÂ¶ÇËù∏ÁâõÊÆºÁöÑËû∫ÊóãÂΩ¢ÁãÄ

```python
# Ë®≠ÂÆöÂèÉÊï∏ÁØÑÂúç
t = np.linspace(0, 4 * np.pi, 1000)

# Ë®≠ÂÆöËù∏ÁâõÊõ≤Á∑öÁöÑÂçäÂæëÂáΩÊï∏
r = np.exp(0.2 * t)

# ËΩâÊèõÁÇ∫Áõ¥ËßíÂ∫ßÊ®ô
x = r * np.cos(t)
y = r * np.sin(t)

# Áπ™Ë£ΩËù∏ÁâõÊõ≤Á∑ö
plt.figure(figsize=(6,6))
plt.plot(x, y, color='orange', linewidth=2)

# Ë®≠ÂÆöÊ®ôÈ°å
plt.title("Snail Spiral", fontsize=14)
plt.xticks([])
plt.yticks([])
plt.grid(False)

# È°ØÁ§∫ÂúñÂΩ¢
plt.show()
```
